
<h2 align='center'>Seaborn: Statistical Data Visualization</h2>


I used **tips** and another very famous dataset **titanic** for this task. These datasets are included in the seaborn's built-in dataset library.<br>
And did some exploratory data analysis using seaborn and present useful plots for data visualization!<br>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.set_style ('white')
titanic= sns.load_dataset('titanic')
titanic
titanic.head()
```

```python
tips= sns.load_dataset('tips')
tips
tips.head()
```

```python
iris= sns.load_dataset('iris')
iris
iris.head()
```

**Creating the box plot Using tips dataset to find how much tip they got on specific days based on their Gender** 

```python
sns.boxplot(x='day', y='tip', hue='sex', data=tips, palette='viridis')
plt.show()
```

**Creating the joint plot Using tips dataset to find the ratio of their amount of tip according to customers total bills.** 

```python
sns.jointplot(data=tips, x='total_bill',y='tip', kind='hex')
plt.show()
```

**Creating the distplot Using titanic dataset and use kde to show only histogram with green color.**

```python
sns.distplot(titanic['fare'], kde= False, color='green', bins=30)
plt.show()
```

**Creating the boxplot below using titanic dataset to find age and their classes in term of dead or alive.** 

```python
sns.boxplot(data=titanic, x="class", y="age", hue="alive",palette='coolwarm')
plt.show()
```

**Creating the joint plot using titanic dataset to find fare over their ages.** 

```python
sns.jointplot(data=titanic, x='fare', y='age')
plt.show()
```

**Creating the box plot Using titanic dataset using color palette on different classes.** 

```python
sns.boxplot(data=titanic, x="class", y="age", palette='rainbow')
plt.show()
```

**Ploting how many male and female were on titanic? And Separating them based on their class.**

```python
sns.countplot(data=titanic, x="sex", hue="class", palette=['#4C72B0', '#55A868', '#C44E52']) 
plt.show()
```

**Creating the heatmap Using titanic dataset.** 

```python
titanic2 = titanic[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'adult_male', 'alone']]
tic = titanic2.corr()
sns.heatmap(tic, annot=True, cmap="coolwarm", center=0, linewidths=.5)
plt.title('titanic.corr()')
plt.show()

#Beacuse of seaborn version problem, values are not appeared on the other text boxes. Thank You.
```
